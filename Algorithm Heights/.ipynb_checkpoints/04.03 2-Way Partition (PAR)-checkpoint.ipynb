## Problem 04.03
A partition procedure is an essential part of the *Quick Sort* algorithm, the subject of one of the following problems. Its main goal is to put the first element of a given array to its proper place in a sorted array. It can be implemented in linear time, by a single scan of a given array. Moreover, it is not hard to come up with an in-place algorithm.

![](http://rosalind.info/media/problems/par/partition2.png "Sketch of the problem")

* **Given**: A positive integer $n\leq10^5$ and an array $A[1\dots n]$ of integers from $−10^5$ to $10^5$.
* ** Return**: A permuted array $B[1\dots n]$ such that it is a permutation of $A$ and there is an index $1\leq q\leq n$ such that $B[i]\leq A[1]$ for all $1\leq i\leq q−1$, $B[q]=A[1]$, and $B[i]>A[1]$ for all $q+1\leq i\leq n$.

```python
n, A = open('rosalind_par.txt','r').read().strip().split('\n')
A = list(map(int, A.split(' ')))
reference = A[0]

lower, upper, equals = [], [], []
for i in range(1, len(A)):
    if A[i] < reference:
        lower.append(A[i])      
    elif A[i] > reference:
        upper.append(A[i])
    else:
        equals.append(A[i])
A = lower+equals+[reference]+upper

print(*A)
```
