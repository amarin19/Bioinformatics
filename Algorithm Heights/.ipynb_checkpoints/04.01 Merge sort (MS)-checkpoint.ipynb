## Problem 04.01

The problem of sorting a list of numbers lends itself immediately to a divide-and-conquer strategy: split the list into two halves, recursively sort each half, and then merge the two sorted sublists (recall the problem *MER*).

 * **Given**: A positive integer $n\leq10^5$ and an array $A[1\dots n]$ of integers from $-10^5$ to $10^5$.
 * **Return**: A sorted array $A[1\dots n]$.

```python
f = open('rosalind_ms.txt', 'r').read().strip().split('\n')
n = int(f[0])
A = list(map(int, f[1].split(' ')))

def merge(A,B):
    C = []
    while A and B:
        if A[0] < B[0]:
            C += [A.pop(0)]
        else:
            C += [B.pop(0)]
    C += A+B
    return C

def mergeSort(A):
    if len(A) > 1:
        A = merge(mergeSort(A[:len(A)//2]), mergeSort(A[len(A)//2:]))
    return A

print(*mergeSort(A))
```
