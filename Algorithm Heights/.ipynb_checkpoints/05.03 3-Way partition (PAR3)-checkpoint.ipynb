## Problem 05.03

This problem is very similar to PAR, but now the goal is to partition an input array more carefully.

![](http://rosalind.info/media/problems/par3/partition3.png "Sketch of the problem")

* **Given**: A positive integer $n\leq10^5$ and an array $A[1\dots n]$ of integers from $−10^5$ to $10^5$.
* ** Return**: A permuted array $B[1\dots n]$ such that it is a permutation of $A$ and there is an index $1\leq q\leq r\leq n$ such that $B[i]< A[1]$ for all $1\leq i\leq q−1$, $B[q]=A[1]$ for all $q\leq i\leq r$, and $B[i]>A[1]$ for all $r+1\leq i\leq n$.

```python
n, A = open('rosalind_par3.txt','r').read().strip().split('\n')
A = list(map(int, A.split(' ')))
reference = A[0]

lower, upper, equals = [], [], []
for i in range(1, len(A)):
    if A[i] < reference:
        lower.append(A[i])      
    elif A[i] > reference:
        upper.append(A[i])
    else:
        equals.append(A[i])
A = lower+equals+[reference]+upper

print(*A)
```
