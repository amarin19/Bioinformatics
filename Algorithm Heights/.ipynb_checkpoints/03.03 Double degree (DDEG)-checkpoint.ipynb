## Problem 03.03
Source: Algorithms by Dasgupta, Papadimitriou, Vazirani. McGraw-Hill. 2006.
* **Given**: A simple graph with $n\leq10^3$ vertices in the edge list format.
* **Return**: An array $D[1\dots n]$ where $D[i]$ is the sum of the degrees of $i$'s neighbors.

```python
import networkx as nx

G = nx.read_edgelist('rosalind_ddeg.txt', nodetype=int, data=False, delimiter=" ")

D = {}
for u in G.nodes():
    s = 0
    for v in G[u]:
        s += G.degree(v)
    D[u] = s

D = sorted(D.items(), key=lambda x:x[0])
__, degs = zip(*D)
print(' '.join(map(str,degs)))
```
