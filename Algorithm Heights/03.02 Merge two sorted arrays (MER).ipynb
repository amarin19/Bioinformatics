## Problem 03.02
The merging procedure is an essential part of *“Merge Sort”* (which is considered in one of the next problems).

 * **Given**: A positive integer $n\leq10^5$ and a sorted array $A[1\dots n]$ of integers from $−10^5$ to $10^5$, a positive integer $m\leq10^5$ and a sorted array $B[1\dots m]$ of integers from $−10^5$ to $10^5$.
 * **Return**: A sorted array $C[1\dots n+m]$ containing all the elements of $A$ and $B$.

*Hint*: the very first element of $C$ is either $A[1]$ or $B[1]$, whichever is smaller. The rest of $C$ can then be constructed recursively.

```python
f = open('rosalind_mer.txt', 'r').read().strip().split('\n')
n, m = int(f[0]), int(f[2])
A, B = list(map(int, f[1].split(' '))), list(map(int, f[-1].split(' ')))

C = []
while A and B:
    if A[0] < B[0]:
        C += [A.pop(0)]
    else:
        C += [B.pop(0)]
C += A+B
print(C)
```
