## Problem 04.04

A binary heap is a binary tree based data structure that is often used to implement priority queues. Binary heaps, in turn, can be easily implemented using an array if the underlying tree is a complete binary tree. The tree nodes have a natural ordering: row by row, starting at the root and moving left to right within each row. If there are $n$ nodes, this ordering specifies their positions $1,2,\dots,n$ within the array. Moving up and down the tree is easily simulated on the array, using the fact that node number $j$ has parent $⌈j/2⌉$ and children $2j$ and $2j+1$.

The goal of this problem is to build a heap from the given array. For this, go from the end to the beginning of a given array and let each element "bubble up".

* **Given**: A positive integer $n\leq10^5$ and array $A[1\dots n]$ of integers from $-10^5$ to $10^5$.
* **Return**: A permuted array $A$ satisfying the binary max heap property: for any $2\leq i\leq n$, $A[⌊i/2⌋]≥A[i]$.

```python
f=open("rosalind_hea.txt", "r")
c=0
l=[]
n=""
l = {a : "" for a in range(0,100000)}
for i in f.readlines():
    if c==0:
        n=i.replace("\n", "")
        c+=1
    else:
        for j in i.split(" "):
            l[c]=int(j.replace("\n", ""))
            c+=1

dele=[]
for i in l:
    if l[i]=="":
        dele.append(i)
for i in dele:
    del l[i]
n=len(l)

for i in range(1, len(l)):
    if l[i]==max(l):
        a=l[1]
        l[1]=l[i]
        l[i]=a
        break
def check(l, i):
    if l[i//2]<l[i]:
        a=l[i]
        l[i]=l[i//2]
        l[i//2]=a
    return l
def overall_check(l):
    for i in range(2, n+1):
        l=check(l, i)
    for i in range(2, n+1):
        if l[i//2]<l[i]:
            overall_check(l)
overall_check(l)
for i in l:
    print(l[i], end = ' ')
```

```python
def BuildMaxHeap (A):
    heap_length = len(A)
    for ind in range(len(A)//2):
        ind = len(A)//2-1-ind
        MaxHeapify(A, ind, heap_length)
    return A

def MaxHeapify (A, i, heap_length):
    left = 2*i
    right = 2*i + 1
    largest = i
    
    if left <= heap_length and A[left] > A[largest]:
        largest = left
    elif right <= heap_length and A[right] > A[largest]:
        largest = right
    elif largest != i:
        A[i], A[largest] = A[largest], A[i]
        MaxHeapify(A, largest, heap_length)
    return A
        
N, A = open('prova.txt', 'r').read().strip().split('\n')
N, A = int(N), list(map(int, A.split(' ')))

print(A)
print(*BuildMaxHeap(A))
```
