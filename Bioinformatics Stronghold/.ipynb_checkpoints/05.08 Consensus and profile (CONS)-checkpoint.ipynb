```python
bases = ['A', 'C', 'G', 'T']
f = open('rosalind_cons.txt', 'r').read().split('>')[1:]
seqs = []
for i in range(len(f)):
    f[i] = f[i].replace('\n','')
    for j in range(len(f[i])):
        if f[i][j-1].isdigit() == True and f[i][j].isdigit() == False:
            seqs.append(f[i][j:])
            break
```

```python
As, Cs, Gs, Ts = ['A:'], ['C:'], ['G:'], ['T:']
consensus = []
for j in range(len(seqs[0])):
    lletra = []
    nmax = []
    for i in range(len(seqs)):
        lletra.append(seqs[i][j])
    for base in bases:
        nmax.append(lletra.count(base))
    consensus.append(bases[nmax.index(max(nmax))])
    As.append(nmax[0])
    Cs.append(nmax[1])
    Gs.append(nmax[2])
    Ts.append(nmax[3])

print(''.join(i for i in consensus))
print(*As)
print(*Cs)
print(*Gs)
print(*Ts)
```
