```python
f = open('rosalind_grph.txt', 'r').read().split('>')[1:]
data = {}
for i in range(len(f)):
    f[i] = f[i].replace('\n','')
    for j in range(len(f[i])):
        if f[i][j-1].isdigit() == True and f[i][j].isdigit() == False:
            data[f[i][0:j]] = f[i][j:]
            break
            
for i in data:
    for j in data:
        if i == j:
            continue
        else:
            if data[i][-3:] == data[j][:3]:
                print('{} {}'.format(i, j))
            else:
                continue
```
