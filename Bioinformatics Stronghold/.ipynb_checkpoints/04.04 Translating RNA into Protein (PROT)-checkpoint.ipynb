```python
#Create the dictionary:
bases = ['U', 'C', 'A', 'G']
aa = 'FFLLSSSSYY**CC*WLLLLPPPPHHQQRRRRIIIMTTTTNNKKSSRRVVVVAAAADDEEGGGG'
codons = [a+b+c for a in bases for b in bases for c in bases]
codon_table = dict(zip(codons, aa))

#Preinitialize variables:
mRNA = open('rosalind_prot.txt', 'r').read().strip()
CDS_of_mRNA = ''
prot = ''

#Find the start (i.e. the Methionine codon) of de coding sequence, CDS:
for i in range(len(mRNA)):
    if codon_table[mRNA[i:i+3]] == 'M':
        CDS_of_mRNA = mRNA[i:]
        break

#Translate the CDS into a protein:
for i in range(0,len(CDS_of_mRNA),3):
    codon = codon_table[CDS_of_mRNA[i:i+3]]
    if codon != '*':
        prot += codon
    else:
        print(prot)
        break
```
