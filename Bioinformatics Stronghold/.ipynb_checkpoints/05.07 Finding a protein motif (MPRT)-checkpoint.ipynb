```python
#Bad thing? The import is manual. urllib works too slowly.
import webbrowser

def findmotifs(ID, seq):
    a = []
    for i in range(len(seq)-5):
        if seq[i] == 'N' and seq[i+1] != 'P' and seq[i+3] != 'P' and seq[i+2] in 'ST':
            a.append(i+1)
    return(a)

f = open('rosalind_mprt.txt', 'r').read().strip().split('\n')
solution = []
for i in range(len(f)):
    url = 'http://www.uniprot.org/uniprot/{}.fasta'.format(f[i])
    webbrowser.open(url)
    seq = input('Enter the sequence :').replace(' ','')
    if not findmotifs(f[i], seq):
        continue
    solution += f[i], findmotifs(f[i], seq)

print()
for i in range(int(len(solution)/2)):
    print(solution[2*i])
    print(*solution[2*i+1])
```
