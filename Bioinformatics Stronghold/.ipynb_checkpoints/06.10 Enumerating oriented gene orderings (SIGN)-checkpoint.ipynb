```python
import itertools
n = int(open('rosalind_sign.txt','r').read().strip().split('\n')[0])

def plusAndMinusPermutations(items):
    for p in itertools.permutations(items):
        for signs in itertools.product([-1,1], repeat=len(items)):
            yield [a*sign for a,sign in zip(p,signs)]   

A = list(plusAndMinusPermutations(range(1,n+1)))
print(len(A))
for i in A:
    print(*i)
```
